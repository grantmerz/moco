### Inference

Obtain representations for the first 16 imges in the validation dataset.
Note that we train the model on 64x64 images while in this example (for illustration purposes) we apply the model on the 107x107 images.

```python
t0 = time.time()
with torch.no_grad():
    #datai = data['images'][0:200][:,:,::-1]
    #    reps = model(torch.tensor(data['images'][0:16]).to(torch.cuda.current_device())).cpu().detach().numpy()
    datai = torch.from_numpy(np.flip(data['images'][0:200],axis=(2,3)).copy())
    reps = model(datai.clone().detach()).cpu().detach().numpy()
print(time.time()-t0)
```

```python
reps.shape
```

```python

with torch.no_grad():
    for i in range(1,30):
        datai = torch.from_numpy(np.flip(data['images'][200*i:200*(i+1)],axis=(2,3)).copy())
        repsi = model(datai.clone().detach()).cpu().detach().numpy()
        #repsi = model(torch.tensor(data['images'][200*i:200*(i+1)][:,:,::-1])).cpu().detach().numpy()
        reps = np.concatenate((reps,repsi))
```
